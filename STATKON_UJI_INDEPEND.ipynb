````python
#A. UJI NORMALITAS JARQUE-BERA
from scipy.stats import jarque_bera

# Data
data=[12,15,14,10,18,16,13,17,11,14]

#menghitung banyak data
n=len(data)

#menghitung mean
mean=sum(data)/n

#membuat list kosong dan loop menghitung selisih nilai observasi dengan rata-rata
selisih=[]
for x in data:
  selisih.append(x-mean)

#menghitung momen ke-2
m2 = 0
for d in selisih:
  m2 += d**2
m2 = m2/n

#m3
m3=0
for d in selisih:
  m3 += d**3
  m3=m3/n

#m4
m4=0
for d in selisih:
  m4+=d**4
m4=m4/n

#menghitung skewness (kemiringan)
skewness = m3/(m2**1.5)

#menghitung nilai kurtosis (keruncingan)
kurtosis = m4 /  (m2**2)

#masukkan rumus pada Jarque Bera
jb = (n-6) * (skewness**2 + ((kurtosis-3)**2) / 4)

#nilai eksak chi square
chi_square_critical = 5.991

#keputusan
if jb > chi_square_critical:
  print ("keputusan = Tolak H0: data tidak berdistribusi normal")
else:
  print ("keputusan = Gagal Tolak H0: data berdistribusi normal")
````

````python
#B. KOLMOGOROV SMIRNOV
import math

def kolmogorov_smirnov_test(data):

    n = len(data)

    data.sort()

    mean = sum(data) / n

    jumlah = 0
    for x in data:
        jumlah += (x - mean) ** 2

    std = (jumlah / (n - 1)) ** 0.5

    D = 0

    print("Data Terurut :", data)
    print("Mean =", mean)
    print("Std Dev =", std)

    for i in range(n):

        x = data[i]

        z = (x - mean) / std

        ft = 0.5 * (1 + math.erf(z / math.sqrt(2)))

        fo = (i + 1) / n

        selisih = abs(fo - ft)

        if selisih > D:
            D = selisih

    d_tabel = 1.36 / math.sqrt(n)

    print("D hitung =", D)
    print("D tabel =", d_tabel)

    if D > d_tabel:
        print("Keputusan : Tolak H0")
        print("Data tidak berdistribusi normal")
    else:
        print("Keputusan : Gagal Menolak H0")
        print("Data berdistribusi normal")

#soal
data = [12,15,14,10,18,16,13,17,11,14]

kolmogorov_smirnov_test(data)
````

# Uji Normalitas Kolmogorov-Smirnov (K-S)

## Pengertian

Uji Kolmogorov-Smirnov (K-S) merupakan metode statistik yang digunakan untuk menguji apakah suatu sampel data mengikuti distribusi tertentu, terutama distribusi normal. Uji ini dilakukan dengan membandingkan distribusi kumulatif empiris dari data sampel dengan distribusi kumulatif teoritis yang diharapkan.

Uji Kolmogorov-Smirnov termasuk salah satu metode pengujian normalitas yang sering digunakan karena mampu mengukur perbedaan antara distribusi data aktual dengan distribusi teoritis secara langsung. Semakin kecil perbedaan antara kedua distribusi tersebut, maka semakin besar kemungkinan data berdistribusi normal.

Pada pengujian ini digunakan dua hipotesis sebagai berikut:

$$
H_0 : \text{Data berdistribusi normal}
$$

$$
H_1 : \text{Data tidak berdistribusi normal}
$$

Keputusan pengujian dilakukan dengan membandingkan nilai statistik Kolmogorov-Smirnov ($D$) dengan nilai kritis pada tabel Kolmogorov-Smirnov. Jika nilai $D$ hitung lebih besar daripada nilai $D$ tabel, maka hipotesis nol ditolak sehingga data dinyatakan tidak berdistribusi normal. Sebaliknya, jika nilai $D$ hitung lebih kecil atau sama dengan nilai $D$ tabel, maka hipotesis nol gagal ditolak dan data dianggap berdistribusi normal.

## Rumus-Rumus

### Distribusi Kumulatif Empiris

Distribusi kumulatif empiris dihitung menggunakan rumus:

$$
F_o(x)=\frac{i}{n}
$$

Keterangan:

- $F_o(x)$ = distribusi kumulatif empiris
- $i$ = urutan data setelah diurutkan
- $n$ = jumlah data

### Standarisasi Data (Z-Score)

Sebelum menghitung distribusi teoritis, setiap data diubah ke bentuk standar menggunakan rumus:

$$
z=\frac{x-\bar{x}}{s}
$$

Keterangan:

- $x$ = nilai data
- $\bar{x}$ = rata-rata data
- $s$ = standar deviasi

### Statistik Kolmogorov-Smirnov

Nilai statistik Kolmogorov-Smirnov dihitung dengan mencari selisih terbesar antara distribusi empiris dan distribusi teoritis.

$$
D=\max |F_o(x)-F_t(x)|
$$

Keterangan:

- $D$ = statistik Kolmogorov-Smirnov
- $F_o(x)$ = distribusi kumulatif empiris
- $F_t(x)$ = distribusi kumulatif teoritis
- $\max$ = nilai maksimum dari seluruh selisih

# Uji Homogenitas Levene

## Pengertian

Uji Levene merupakan metode statistik yang digunakan untuk menguji apakah dua atau lebih kelompok data memiliki varians yang sama (homogen). Uji ini sering digunakan sebagai alternatif Uji F karena lebih tahan terhadap pelanggaran asumsi normalitas. Oleh karena itu, Uji Levene banyak digunakan sebagai syarat awal sebelum melakukan analisis lanjutan seperti Uji T dan ANOVA.

Prinsip dasar Uji Levene adalah mengukur seberapa besar penyimpangan setiap data terhadap rata-rata kelompoknya. Jika penyimpangan antar kelompok relatif sama, maka varians kelompok dianggap homogen.

Pada pengujian ini digunakan hipotesis sebagai berikut:

$$
H_0 : \text{Varians seluruh kelompok sama (homogen)}
$$

$$
H_1 : \text{Varians seluruh kelompok berbeda (tidak homogen)}
$$

Keputusan pengujian dilakukan dengan membandingkan nilai statistik Levene dengan nilai kritis distribusi F. Jika nilai statistik Levene lebih besar daripada nilai F tabel, maka hipotesis nol ditolak. Sebaliknya, jika nilai statistik Levene lebih kecil atau sama dengan nilai F tabel, maka hipotesis nol gagal ditolak.

## Rumus-Rumus

### Selisih Absolut terhadap Rata-Rata Kelompok

Nilai absolut setiap data terhadap rata-rata kelompok dihitung dengan rumus:

$$
Z_{ij}=|X_{ij}-\bar{X}_i|
$$

Keterangan:

- $X_{ij}$ = data ke-j pada kelompok ke-i
- $\bar{X}_i$ = rata-rata kelompok ke-i
- $Z_{ij}$ = nilai selisih absolut

### Statistik Levene

$$
W=
\frac{(N-k)}
{(k-1)}
\cdot
\frac{\sum_{i=1}^{k} n_i(Z_i-Z)^2}
{\sum_{i=1}^{k}\sum_{j=1}^{n_i}(Z_{ij}-Z_i)^2}
$$

Keterangan:

- $W$ = statistik Levene
- $N$ = jumlah seluruh data
- $k$ = jumlah kelompok
- $n_i$ = jumlah data kelompok ke-i
- $Z_i$ = rata-rata nilai absolut kelompok ke-i
- $Z$ = rata-rata keseluruhan nilai absolut

````python
#A. UJI CHI SQUARE
from scipy.stats import chi2

def uji_chi_square(data, alpha=0.05):

    jumlah_baris = len(data)
    jumlah_kolom = len(data[0])

    total_baris = []
    for baris in data:
        total_baris.append(sum(baris))

    total_kolom = []

    for j in range(jumlah_kolom):
        total = 0
        for i in range(jumlah_baris):
            total += data[i][j]

        total_kolom.append(total)

    grand_total = sum(total_baris)

    harapan = []

    for i in range(jumlah_baris):

        baris = []

        for j in range(jumlah_kolom):

            eij = (total_baris[i] * total_kolom[j]) / grand_total

            baris.append(eij)

        harapan.append(baris)

    chi_square = 0

    for i in range(jumlah_baris):
        for j in range(jumlah_kolom):

            O = data[i][j]
            E = harapan[i][j]

            chi_square += ((O - E)**2) / E

    df = (jumlah_baris - 1) * (jumlah_kolom - 1)

    chi_tabel = chi2.ppf(1 - alpha, df)

    if chi_square > chi_tabel:
        keputusan = "Tolak H0: terdapat hubungan"
    else:
        keputusan = "Gagal menolak H0: tidak terdapat hubungan"

    print("jumlah_baris =", jumlah_baris)
    print("jumlah_kolom =", jumlah_kolom)
    print("grand_total =", grand_total)
    print("harapan =", harapan)
    print("chi_square =", chi_square)
    print("chi_tabel =", chi_tabel)
    print("keputusan =", keputusan)

    #soal
    data = [
    [30,20],
    [15,35]
]

uji_chi_square(data)
````

````python
#FISHER'S EXAVT TEST
import math

def fisher_exact_test(data):

    a = data[0][0]
    b = data[0][1]
    c = data[1][0]
    d = data[1][1]

    n = a + b + c + d

    pembilang = (
        math.factorial(a + b)
        * math.factorial(c + d)
        * math.factorial(a + c)
        * math.factorial(b + d)
    )

    penyebut = (
        math.factorial(a)
        * math.factorial(b)
        * math.factorial(c)
        * math.factorial(d)
        * math.factorial(n)
    )

    p_value = pembilang / penyebut

    print("a =", a)
    print("b =", b)
    print("c =", c)
    print("d =", d)

    print("n =", n)
    print("P-value =", p_value)

    alpha = 0.05

    if p_value < alpha:
        print("Keputusan : Tolak H0")
        print("Terdapat hubungan")
    else:
        print("Keputusan : Gagal Menolak H0")
        print("Tidak terdapat hubungan")


data = [
    [30,20],
    [15,35]
]

fisher_exact_test(data)
````

# Fisher's Exact Test

## Pengertian

Fisher's Exact Test merupakan metode statistik yang digunakan untuk menguji hubungan antara dua variabel kategorik yang disajikan dalam tabel kontingensi berukuran 2×2. Uji ini sangat cocok digunakan ketika ukuran sampel relatif kecil sehingga asumsi yang diperlukan pada Uji Chi-Square tidak terpenuhi.

Metode ini menghitung probabilitas secara tepat (exact probability) berdasarkan kombinasi data yang diamati. Oleh karena itu, hasil yang diperoleh lebih akurat untuk sampel kecil dibandingkan menggunakan pendekatan Chi-Square.

Hipotesis yang digunakan adalah sebagai berikut:

$$
H_0 : \text{Tidak terdapat hubungan antara kedua variabel}
$$

$$
H_1 : \text{Terdapat hubungan antara kedua variabel}
$$

Keputusan pengujian dilakukan dengan membandingkan nilai p-value dengan tingkat signifikansi ($\alpha$). Jika p-value lebih kecil dari $\alpha$, maka hipotesis nol ditolak. Sebaliknya, jika p-value lebih besar atau sama dengan $\alpha$, maka hipotesis nol gagal ditolak.

## Rumus-Rumus

Jumlah seluruh data:

$$
n=a+b+c+d
$$

### Rumus Probabilitas Fisher

$$
P=
\frac{(a+b)!(c+d)!(a+c)!(b+d)!}
{a!b!c!d!n!}
$$

Keterangan:

- $a,b,c,d$ = frekuensi pada tabel kontingensi
- $n$ = jumlah seluruh data
- $!$ = operasi faktorial
- $P$ = probabilitas Fisher
